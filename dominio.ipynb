```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funcin
def f(x, y):
    return x*y

# Crear malla
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

# Graficar curvas de nivel
plt.figure()
contours = plt.contour(X, Y, Z, levels=20)
plt.clabel(contours)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Curvas de nivel de f(x,y)=2x+4y")
plt.axhline(0)
plt.axvline(0)
plt.gca().set_aspect('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Superficie del cono elptico
def compute_surface(y_max):
    theta = np.linspace(0, 2*np.pi, 200)
    y = np.linspace(-y_max, y_max, 200)
    theta, y = np.meshgrid(theta, y)

    # Radio elptico
    x = np.sqrt(3/4) * np.abs(y) * np.cos(theta)
    z = np.sqrt(3/2) * np.abs(y) * np.sin(theta)

    return x, y, z


# Crear figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.subplots_adjust(bottom=0.25)

y_max_init = 3

x, y, z = compute_surface(y_max_init)
surface = ax.plot_surface(x, y, z)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("4x - 3y + 2z = 0")

# Slider
ax_slider = plt.axes([0.25, 0.1, 0.5, 0.03])
slider = Slider(ax_slider, 'Altura', 1, 6, valinit=y_max_init)

def update(val):
    ax.clear()
    y_max = slider.val
    x, y, z = compute_surface(y_max)
    ax.plot_surface(x, y, z)
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.set_title("4x - 3y + 2z = 0")
    
    fig.canvas.draw_idle()

slider.on_changed(update)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# ---------- Superficie ----------
def compute_surface(y_max):
    theta = np.linspace(0, 2*np.pi, 200)
    y = np.linspace(-y_max, y_max, 200)
    theta, y = np.meshgrid(theta, y)

    x = np.sqrt(3/4) * np.abs(y) * np.cos(theta)
    z = np.sqrt(3/2) * np.abs(y) * np.sin(theta)

    return x, y, z


# ---------- Crear figura ----------
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
plt.subplots_adjust(bottom=0.25)

y_max_init = 3
x, y, z = compute_surface(y_max_init)

surf = ax.plot_surface(
    x, y, z,
    alpha=0.6,          #  transparencia
    edgecolor='none'
)

# ---------- Centrar ejes ----------
def center_axes(ax, limit):
    ax.set_xlim(-limit, limit)
    ax.set_ylim(-limit, limit)
    ax.set_zlim(-limit, limit)
    ax.set_box_aspect([1,1,1])  # proporcin igual

    # Dibujar ejes coordenados
    ax.plot([-limit, limit], [0,0], [0,0])
    ax.plot([0,0], [-limit, limit], [0,0])
    ax.plot([0,0], [0,0], [-limit, limit])

center_axes(ax, y_max_init)

ax.set_title("Cono Elptico: 4x - 3y + 2z = 0")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# ---------- Slider ----------
ax_slider = plt.axes([0.25, 0.1, 0.5, 0.03])
slider = Slider(ax_slider, 'Altura', 1, 6, valinit=y_max_init)

def update(val):
    ax.cla()
    y_max = slider.val
    x, y, z = compute_surface(y_max)

    ax.plot_surface(
        x, y, z,
        alpha=0.6,
        edgecolor='none'
    )

    center_axes(ax, y_max)

    ax.set_title("Cono Elptico: 4x - 3y + 2z = 0")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")

    fig.canvas.draw_idle()

slider.on_changed(update)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear malla
x = np.linspace(-5, 5, 600)
y = np.linspace(-5, 5, 600)
X, Y = np.meshgrid(x, y)

# Definir funcin evitando divisin por 0
R2 = X**2 + Y**2
Z = np.zeros_like(X)
mask = R2 != 0
Z[mask] = 2 * X[mask] * Y[mask] / R2[mask]

# Dibujar curvas de nivel
plt.figure(figsize=(6,6))
niveles = np.linspace(-1, 1, 11)
contours = plt.contour(X, Y, Z, levels=niveles)

plt.clabel(contours, inline=True, fontsize=8)
plt.axhline(0)
plt.axvline(0)

plt.gca().set_aspect('equal', 'box')
plt.title("Curvas de nivel de f(x,y) = 2xy/(x+y)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear malla
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

R2 = X**2 + Y**2
Z = np.zeros_like(X)

# Evitar divisin por 0
mask = R2 != 0
Z[mask] = 2 * X[mask] * Y[mask] / R2[mask]

# Crear figura
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(
    X, Y, Z,
    cmap='coolwarm',
    edgecolor='none',
    alpha=0.9
)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.set_title("Superficie f(x,y) = 2xy / (x + y)")

ax.set_zlim(-1, 1)
ax.set_box_aspect([1,1,0.6])

fig.colorbar(surface, shrink=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear malla
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

# Definir funcin
Z = np.exp(-(X**2 + Y**2))

# Crear figura
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(
    X, Y, Z,
    cmap='viridis',
    edgecolor='none',
    alpha=0.9
)

# Etiquetas
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("Superficie z = e^{-(x^2+y^2)}")

# Proporcin visual agradable
ax.set_box_aspect([1,1,0.6])

fig.colorbar(surface, shrink=0.6)

plt.show()
```
