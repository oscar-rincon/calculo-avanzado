```python
# 2D contour plot of q(x,y) = 34x^2 + 24xy + 41y^2

# Create grid
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Define function
Z = 34*X**2 + 24*X*Y + 41*Y**2

# Create contour plot
plt.figure()
plt.contourf(X, Y, Z, levels=20)
plt.colorbar(label='q(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
